```python
df =df.drop(columns = ['Acronym','Study Documents'])
```

```python
df['Result First Posted'] = df['Results First Posted'].fillna('Unknown',inplace = True)
```

```python
columns = list(df)
columns
```

```python
(df[columns[1:6]]==0).sum()
```

```python
df[columns[1:6]] = df[columns[1:6]].replace(0,np.nan) # identfying null values
```

```python
df.isnull().sum()
```

```python
col = df['Phases']
col = col.fillna(col.mode()[0], inplace = True) #replacing the null values with mode
```

```python
df =df.drop(columns = ['Result First Posted'])
```

```python
print(df_new['Phases'].value_counts())
df['Phases'].value_counts().plot(kind = 'bar', title = 'Distribution of Phases')
```

the graph displays the distribution of clinical trials asross various phases.
X axis: Different phases
Y axis : count

```python
status_phase = pd.crosstab(df['Status'], df['Phases'])
print(status_phase)
status_phase.plot(kind='bar', stacked=True, title='Status vs.Phases')
plt.show()
```

recruititng : has the highest number of trials mostly in early phases

```python
conditions_outcomes = df_new.groupby('Conditions')['Outcome Measures'].apply(lambda x: ', '.join(x)).reset_index()
print(conditions_outcomes)
```

```python
df['Start Date'] = pd.to_datetime(df['Start Date'],errors='coerce')
df['Primary Completion Date'] = pd.to_datetime(df['Primary Completion Date'], errors='coerce')

# Plot the number of trials started over time
df['Start Date'].dt.to_period('M').value_counts().sort_index().plot(kind='line', title='Trials Started Over Time')
```

2019 : masive spike in the no.of trials depicting mobilization of clinical research due to emergence of COVID-19

```python
funding_counts = df['Funded Bys'].value_counts()
funding_counts = funding_counts.head(10)
plt.figure(figsize=(20, 6))
plt.bar(funding_counts.index, funding_counts.values, color=['blue', 'green', 'orange'])
plt.title('Studies by Funding Source')
plt.xlabel('Funding Source')
plt.ylabel('Number of Studies')
plt.show()
```

shows the number of studies by different funding sources.
Others being the heighest covering the vast majority of funding researches


```python
avg_enrollment = df.groupby('Phases')['Enrollment'].mean()

plt.figure(figsize=(8, 6))
plt.barh(avg_enrollment.index, avg_enrollment.values, color='lightblue')
plt.title('Average Enrollment by Phase')
plt.xlabel('Average Enrollment')
plt.ylabel('Phase')
```

x-axis: avg. no. of participants enrolled
y-axis: list of phases
the heighest no. of enrolments are in Not Applicable  phase
